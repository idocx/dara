```python
import matplotlib.pyplot as plt
import numpy as np


def gaussian(x, mu, sig, scale):
    return np.exp(-np.power(x - mu, 2.0) / (2 * np.power(sig, 2.0))) * scale


# set background color to white
plt.rcParams["figure.facecolor"] = "white"

# remove frame
ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_visible(False)

# remove ticks
plt.tick_params(
    axis="both",
    which="both",
    bottom=False,
    top=False,
    left=False,
    right=False,
    labelbottom=False,
    labelleft=False,
)

x = np.linspace(-10, 3, 500)

plt.fill(
    x,
    np.clip(
        gaussian(x, 0, 0.6, 0.8)
        + gaussian(x, -7, 0.8, 0.5)
        + gaussian(x, -3.5, 1.2, 1),
        0.05,
        None,
    ),
    lw=0,
    color="#ECB365",
    alpha=0.4,
)

plt.fill(
    x + 0.7,
    np.clip(gaussian(x, -0.0, 0.6, 0.8), 0.05, None) - 0.04,
    lw=0,
    color="#00A9FF",
    alpha=0.5,
)
plt.fill(
    x + 0.7,
    np.clip(gaussian(x, -3.6, 1.2, 1), 0.05, None) - 0.04,
    lw=0,
    color="#069A8E",
    alpha=0.5,
)
plt.fill(
    x + 0.7,
    np.clip(gaussian(x, -7, 0.8, 0.48), 0.05, None) - 0.04,
    lw=0,
    color="#B983FF",
    alpha=0.5,
)

plt.text(
    4,
    0.6,
    "DARA  ",
    fontsize=96,
    color="#1B4242",
    ha="left",
    va="center",
    fontdict={"weight": "bold"},
)
plt.text(
    4.2,
    0.2,
    "Data-driven automated Rietveld analysis\nfor phase search and refinement",
    fontsize=28,
    color="#5C8374",
    ha="left",
    va="center",
)

plt.savefig("dara.jpg", dpi=300, bbox_inches="tight")
```
